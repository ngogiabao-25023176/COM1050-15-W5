```python
def songuyento(n):
    if n<2:
        return False
    
    for i in range(2,n):
        if n%i==0:
            return False
    return True

n=int(input())
if songuyento(n):
    print ('yes')
else:
    print ('no')
```

```python
def giaithua(n):
    tich=1
    for i in range(1,n+1):
        tich*=i
    return tich

n=int(input())
print (giaithua(n))
```

```python
def maytinh(a,b,dau):
    if dau=="-":
        print (a-b)
    elif dau=="+":
        print (a+b)
    elif dau=="/":
        print (a/b)
    elif dau=="*":
        print (a*b)
    else:
        print ('nhập dấu không hợp lệ')
        
a,b=map(int,input().split())
dau=input().strip()
maytinh(a,b,dau)
```

```python
def hamming(x,y):
    dem=0
    while x>0 or y>0:
        if x%2!=y%2:
            dem += 1
        x //= 2
        y //= 2
    return dem

a,b=map(int,input().split())
print (hamming(a,b))
```

```python
def tong(n):
    dem=0
    while n!=0:
        tmp=n%10
        dem+=tmp
        n//=10
    return dem

n=int(input())
print (tong(n))
```

```python
def dangcau(a, b):
    map_a_b = {}
    map_b_a = {}

    for i in range(len(a)):
        x = a[i]
        y = b[i]

        if x in map_a_b and map_a_b[x] != y:
            return False
        
        if y in map_b_a and map_b_a[y] != x:
            return False

        map_a_b[x] = y
        map_b_a[y] = x

    return True

a=input()
b=input()
print (dangcau(a,b))
```
